```python
df3.head()
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    else:
        one = pd.get_dummies(df3[col], drop_first=True)
        df3.drop(df3[[col]], axis=1,inplace=True)
        df3=pd.concat([df3,one], axis=1)
```
